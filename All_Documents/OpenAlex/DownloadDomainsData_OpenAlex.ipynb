```python
def get_country_code(cname):
  for continent in COUNTRIES_BY_CONTINENT:
    for country in COUNTRIES_BY_CONTINENT[continent]:
      if country["name"] == cname:
        return country["country_code"]
  return None

get_country_code('Turkey')
```

```python
years = list(range(1980, 2025))
```

```python
import pandas as pd

#Diccionario con una dataframe de produccion Años x Areas por cada unidad
df_units = {}
#Diccionario con la producción anual de los paises en un dataframe (nombre del índice 'Publication Year', columna 'Documents')
df_units_production = {}
#Producción mundial: Año x Areas
df_areas = pd.DataFrame(index=years, columns=areas)
```

```python
# Import time
import time
world_works = Works().group_by("publication_year").get()
time.sleep(0.25)

#Producción del mundo
df_units_production['Baseline for All Items'] = pd.DataFrame(index=years, columns=['Documents'])

for register in world_works:
  year = int(register['key'])
  if year in years:
    # Retrive  world works in the year grouped by field and put it into the dictionary
    works = Works().filter(publication_year=year).group_by("primary_topic.field.id").get()
    df_units_production['Baseline for All Items'].at[year, 'Documents'] = register['count']
    for field in works:
      df_areas.at[year, field['key_display_name']] = field['count']
    print("", end=f"\rYear: {year} %")
    time.sleep(0.25)
```

```python
##  THIS CELL DOWNLOADS ALL DATA
# Import time
import time

#countries
for continent in COUNTRIES_BY_CONTINENT:
  for country in COUNTRIES_BY_CONTINENT[continent]:
    country_code = country["country_code"]
    country_name = country["name"]
    df_units[country_code] = pd.DataFrame(index=years, columns=areas)
    df_units_production[country_code] = pd.DataFrame(index=years, columns=['Documents'])
    #Retrive country works per year
    c_works = Works().filter(authorships={"countries": country_code}).group_by("publication_year").get()
    print(country_name)
    time.sleep(0.25)
    for register in c_works:
      year = int(register['key'])
      if year in years:
        # Retrive  country works in the year grouped by field and put it into the dictionary
        works = Works().filter(publication_year=year).filter(institutions={"country_code": country_code}).group_by("primary_topic.field.id").get()
        df_units_production[country_code].at[year, 'Documents'] = register['count']
        for field in works:
          df_units[country_code].at[year, field['key_display_name']] = field['count']
        print("", end=f"\rYear: {year} %")
        time.sleep(0.25)
    print("")
```
